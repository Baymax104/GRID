```python
from pathlib import Path
```

```python
beauty_data_dir = Path("../../../data/beauty").resolve()
beauty_data_dir
```

```python
items_dir = beauty_data_dir / "items"
items_dir
```

```python
data_files = list(items_dir.iterdir())
len(data_files)
```

```python
import tensorflow as tf
dataset = tf.data.TFRecordDataset(data_files, compression_type="GZIP")
dataset
```

```python
sample = next(iter(dataset))
example = tf.train.Example()
example.ParseFromString(sample.numpy())
sample = example
sample
```

```python
sample.features.feature
```

```python
def infer_feature_type(example_proto: tf.Tensor) -> dict:
    feature_description = {}
    # feature has variable length, RaggedFeature in TF 2.x is recommended
    # VarLenFeature is used for compatibility
    tf_feature_type = tf.io.VarLenFeature
    for key, value in example_proto.items():
        if isinstance(value, tf.train.Feature):
            if value.HasField("bytes_list"):
                feature_description[key] = tf_feature_type(tf.string)
            elif value.HasField("float_list"):
                feature_description[key] = tf_feature_type(tf.float32)
            elif value.HasField("int64_list"):
                feature_description[key] = tf_feature_type(tf.int64)
            else:
                raise ValueError("Unknown feature type")
    return feature_description
```

```python
feature_description = infer_feature_type(sample.features.feature)
feature_description
```

```python
tf.io.parse_example(next(iter(dataset)), feature_description)
```
